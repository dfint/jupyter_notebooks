Merge translations missing in the current versions (51.*) from the old DF version (0.47.05)

```python
from babel.messages import Catalog, pofile
```

```python
source = "for_translation_dwarf-fortress-steam_hardcoded_steam_de.po"
```

```python
reference = "for_use_dwarf-fortress_hardcoded_de.po"
```

```python
output = "output.po"
```

```python
with open(reference, encoding="utf-8") as reference_file:
    catalog = pofile.read_po(reference_file)
    dictionary = {
        message.id: message.string
        for message in catalog
        if message.id and message.string and message.id != message.string
    }
```

```python
dictionary
```

```python
with open(source, encoding="utf-8") as source_file:
    source_catalog = pofile.read_po(source_file)
    output_catalog = Catalog()
    for message in source_catalog:
        translation = message.string or dictionary.get(message.id)

        output_catalog.add(
            id=message.id,
            string=translation,
            locations=message.locations,
            flags=message.flags,
            auto_comments=message.auto_comments,
            user_comments=message.user_comments,
            previous_id=message.previous_id,
            lineno=message.lineno,
            context=message.context,
        )
```

```python
with open(output, "wb") as output_file:
    pofile.write_po(output_file, output_catalog)
```
